## 1) Data Description

### Players Dataset:

```r
players <- read_csv("https://raw.githubusercontent.com/audreyw28/DSCI-100-Project/refs/heads/main/players.csv")
head(players)
```

```r
players_range <- players|>
    summarize(played_hours_range = range(played_hours), 
              age_range = range(Age, na.rm = TRUE))
players_range

players_stats <- players|>
    summarize(played_hours_mean = round(mean(played_hours), 2),
              played_hours_sd = round(sd(played_hours), 2),
              age_mean = round(mean(Age, na.rm = TRUE), 2), 
              age_sd = round(sd(Age, na.rm = TRUE), 2))
players_stats

# sum(is.na(players))
```

Data regarding all the players' information was collected as players signed in to play the game. 

In the `players` dataset, there are:
- 196 observations, each of one player
- 7 variables:

| Variable | Type | Description |
|:---|:---|:---|
| experience | character | experience level of player |
| subscribe | logical | whether the player has subscribed to the game or not |
| hashedEmail | character | hashed email of player |
| played_hours | double | number of hours played (hours)|
| name | character | name of player |
| gender | character | gender of player |
| Age | double | age of player (years) |

`played_hours` has: 
- range of 0 - 223.1
- mean of 5.85
- standard deviation of 28.36

`Age` has:
- range of 9 - 58
- mean of 21.14
- standard deviation of 7.39

Issues exist in this dataset in that: 
- 2 player ages were missing (NA) $ \rightarrow $ decreased accuracy when performing summary statistics for `Age`
- the experience levels were self-declared by players $ \rightarrow $ could be biased

## 2) Questions

Broad question: 
- Question 2: "We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts".
    - Since the team wants to obtain data about how people play games, having a high `played_hours` value means large data contribution.
    - `experience` is self-declared and therefore unreliable.
    - `name`, `hashedEmail`, and `gender` are unlikely to be informative in predicting `played_hours`.
    - `subscribe` may be predictive since players who play longer are often subscribed to the game. 
    - `Age` may be predictive if the game appeals to players of certain ages.

Therefore, my specific question is: 
- Can `Age` and `subscribe` predict `played_hours` in the `players` dataset?   

The `players` dataset will make this regression problem possible by providing sufficient (196) observations of players and their respective `played_hours`, `Age`, and `subscribe`. 

## 3) Exploratory Data Analysis and Visualization

Dataset is tidy; my variables of interest (`played_hours`, `Age`, and `subscribe`) are in separate columns, and each observation is one player.
To cater the data to my question, I will drop NAs and select for my interested variables.

```r
players_refined <- players |>
    select(played_hours, Age, subscribe)|>
    drop_na()

head(players_refined)
```

### Scatter Plots of Number of Hours Played vs Player Age

```r
options(repr.plot.height = 8, repr.plot.width = 8)
trend <- ggplot(players_refined, aes(x = Age, y = played_hours, color = subscribe)) +
    geom_point(alpha = 0.7) +
    scale_color_discrete(labels = c("No", "Yes"),
                         name = "Subscribed or Not") +
    labs(x = "Age (years)", 
         y = "Hours Played (hours)", 
         title = "Scatter Plot #1\nRelationship Between Number of\nHours Played and Player Age") +
    theme(text = element_text(size = 16), 
          legend.text = element_text(size = 14),
         legend.title = element_text(size = 16))
trend

trend_log <- ggplot(players_refined, aes(x = Age, y = played_hours, color = subscribe)) +
    geom_point(alpha = 0.7) +
    scale_color_discrete(labels = c("No", "Yes"),
                         name = "Subscribed or Not") +
    labs(x = "Age (years)", 
         y = "Hours Played (hours)", 
         title = "Scatter Plot #2\nRelationship Between Number of\nHours Played (log10) and Player Age") +
    theme(text = element_text(size = 16), 
         legend.text = element_text(size = 14),
         legend.title = element_text(size = 16)) +
    scale_y_log10(labels = label_comma())+
    scale_x_log10(labels = label_comma())
trend_log
```

Scatter plots of `played_hours` vs `Age` (both regular and logged) show a non-linear relationship. Therefore, we should use the K-nearest-neighbors model to perform regression. By coloring with the subscription status, we see that higher `played_hours` are associated with players being subscribed. 

### Histogram of Distribution of Played Hours by Subscription Status

```r
options(repr.plot.height = 10, repr.plot.width = 10)
played_hours_histogram <- ggplot(players_refined, aes(x = played_hours, fill = subscribe))+
    geom_histogram(position = "identity", binwidth = 5, color = "black") + 
    scale_fill_discrete(labels = c("No", "Yes"),
                         name = "Subscribed or Not") +
    labs(x = "Played Hours (Hours)", 
         y = "Count", 
         title = "Histogram of Played Hours") +
    facet_grid(rows = vars(subscribe)) +
    theme(text = element_text(size = 16), 
         legend.text = element_text(size = 14),
         legend.title = element_text(size = 16))
played_hours_histogram
```

These two histogram plots further reinforce that players with high `played_hours` are likely to be subscribed. Therefore, `subscribe` can be used as a predictor variable. 

### Histogram of Distribution of Player Age

```r
options(repr.plot.height = 8, repr.plot.width = 8)
age_distribution <- ggplot(players_refined, aes(x = Age)) +
    geom_histogram(binwidth = 2, color = "black", fill = "orange") +
    labs(x = "Age (year)", y = "Count", title = "Distribution of Age in the\nPlayers Dataset") +
    theme(text = element_text(size = 16))
age_distribution
```

The `age` histogram shows that the majority of players are 16-17 year olds, and `played_hours` for these ages is widely spread (seen from scatter plot #1 and the calculated SD of 28.36). Therefore, predictions for these ages have high uncertainty. For ages with fewer observations, uncertainty also exists as predicted response will be  biased towards ages 16-17. 

## 4) Methods and Plan

- Model Choice: The relationship between the predictor and response variables is not linear, so **knn regression should be used.**
    - This model will assume that data points similar in `Age` and same in `subscribe` should be similar in `played_hours`.
    - Limitation: cannot accurately predict `played_hours` beyond given range of `Age`

**Some steps I will take in building the model:**
1. Split data into 70% training and 30% (most commonly used proportion in my current experience of predictive analysis)
2. `Age` will be standardized (mean=0, SD=1). The two binary entries for `subscribe` will be converted to 0/1. In this way, the two variables will be comparable. 
3. Perform 5-fold cross validation (most commonly used in this course) on 1-30 neighbors (K range ~1/5 of the training data observations) to tune the model.
4. Select neighbors with the lowest cross-validated RMSE for our final model.
